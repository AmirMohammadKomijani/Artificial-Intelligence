در سه قسمت بعد، برخی اطلاعات دیتاست را چاپ می‌کنیم. هر کدام از توابع زیر، به ما اطلاعات کلی ای در مورد دیتاست مورد نظر می‌دهند.
برای آشنایی بیشتر با هریک از توابع زیر، می‌توانید در اینترنت این توابع را جست و جو کنید.

در این بخش، برای اینکه بتوانیم عملیات‌های احتمالی مدنظرمان را روی دیتاست انجام بدهیم، نیاز است که ابتدا مقادیر ویژگی‌ها را گسسته کنیم.
برای این مهم، ابتدا هر کدام از این ویژگی‌ها را باتوجه به مقادیر که در متد 
describe
به دست آمده بود، به چهار بازه تقسیم می‌کنیم و سپس در بخش بعد، به هر یک از دسته‌ها لیبل متناظر با آن‌ها را نسبت می‌دهیم.

```python
df.loc[(df.SepalLengthCm >= 4.3000) & (df.SepalLengthCm < 5.1000), "SepalLengthCm"] = 0
df.loc[(df.SepalLengthCm >= 5.1000) & (df.SepalLengthCm < 5.8000), "SepalLengthCm"] = 1
df.loc[(df.SepalLengthCm >= 5.8000) & (df.SepalLengthCm < 6.4000), "SepalLengthCm"] = 2
df.loc[(df.SepalLengthCm >= 6.4000) & (df.SepalLengthCm <= 7.9000), "SepalLengthCm"] = 3

df.loc[(df.SepalWidthCm >= 2.0000) & (df.SepalWidthCm < 2.8000), "SepalWidthCm"] = 0
df.loc[(df.SepalWidthCm >= 2.8000) & (df.SepalWidthCm < 3.0000), "SepalWidthCm"] = 1
df.loc[(df.SepalWidthCm >= 3.0000) & (df.SepalWidthCm < 3.3000), "SepalWidthCm"] = 2
df.loc[(df.SepalWidthCm >= 3.3000) & (df.SepalWidthCm <= 4.4000), "SepalWidthCm"] = 3

df.loc[(df.PetalLengthCm >= 1.0000) & (df.PetalLengthCm < 1.600000), "PetalLengthCm"] = 0
df.loc[(df.PetalLengthCm >= 1.600000) & (df.PetalLengthCm < 4.350000), "PetalLengthCm"] = 1
df.loc[(df.PetalLengthCm >= 4.350000) & (df.PetalLengthCm < 5.100000), "PetalLengthCm"] = 2
df.loc[(df.PetalLengthCm >= 5.100000) & (df.PetalLengthCm <= 6.90000), "PetalLengthCm"] = 3

df.loc[(df.PetalWidthCm >= 0.100000) & (df.PetalWidthCm < 0.300000), "PetalWidthCm"] = 0
df.loc[(df.PetalWidthCm >= 0.300000) & (df.PetalWidthCm < 1.300000), "PetalWidthCm"] = 1
df.loc[(df.PetalWidthCm >= 1.300000) & (df.PetalWidthCm < 1.800000), "PetalWidthCm"] = 2
df.loc[(df.PetalWidthCm >= 1.800000) & (df.PetalWidthCm <= 2.50000), "PetalWidthCm"] = 3
```

```python
df.loc[df.SepalLengthCm == 0, "SepalLengthCm"] = "XS"
df.loc[df.SepalLengthCm == 1, "SepalLengthCm"] = "S"
df.loc[df.SepalLengthCm == 2, "SepalLengthCm"] = "M"
df.loc[df.SepalLengthCm == 3, "SepalLengthCm"] = "L"

df.loc[df.SepalWidthCm == 0, "SepalWidthCm"] = "XS"
df.loc[df.SepalWidthCm == 1, "SepalWidthCm"] = "S"
df.loc[df.SepalWidthCm == 2, "SepalWidthCm"] = "M"
df.loc[df.SepalWidthCm == 3, "SepalWidthCm"] = "L"

df.loc[df.PetalLengthCm == 0, "PetalLengthCm"] = "XS"
df.loc[df.PetalLengthCm == 1, "PetalLengthCm"] = "S"
df.loc[df.PetalLengthCm == 2, "PetalLengthCm"] = "M"
df.loc[df.PetalLengthCm == 3, "PetalLengthCm"] = "L"

df.loc[df.PetalWidthCm == 0, "PetalWidthCm"] = "XS"
df.loc[df.PetalWidthCm == 1, "PetalWidthCm"] = "S"
df.loc[df.PetalWidthCm == 2, "PetalWidthCm"] = "M"
df.loc[df.PetalWidthCm == 3, "PetalWidthCm"] = "L"
```
